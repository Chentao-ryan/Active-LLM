```python
import openai
openai.api_key  = ''
```

```python
def get_completion(prompt, model="gpt-3.5-turbo-16k"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
```

```python
import time

def connect_with_api(prompt):
    max_retries = 11
    retries = 0

    while retries < max_retries:
        try:
            response = get_completion(prompt)
            return response
        except Exception as e:
            print(f"An error occurred: {e}")
            retries += 1
            print(f"Retrying... (Attempt {retries} of {max_retries})")
            time.sleep(6)  # Wait for 6 second before retrying

    print("Failed to connect with the API after multiple attempts.")
    return None  # Return None or raise an exception, depending on your requirements
```

```python

i = 0  # Start with index 0
response=[None]*X.size
while i < X.size:
    prompt = f"""
    What is the sentiment of the following movie review,
    and how confident you are about this "sentiment",
    which is delimited with triple backticks?

    Give your answer as a single word, either "positive"
    or "negative" and a single percentage in JSON format delimited with space.

    Review text: '''{X[i]}'''
    """

    response[i] = connect_with_api(prompt)
    if response[i] is not None:
        print(response[i], i)
        i += 1  # Proceed to the next index only if the API call was successful
    else:
        print("Skipping to the next index...")
```
